# What is Matplotlib
- ### Matplotlib is a python package used for 2D graphics
- ### Types of plots
    - #### Bar Graph
    - #### Histogram
    - #### Scatter plot
    - #### Pie plot
    - #### Area plot

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3], [4,5,2])   #Simple graph x and y axis
plt.show()
```

```python
# Giving label and title
import matplotlib.pyplot as plt

x = [5,6,8]
y = [7,3,6]
plt.plot(x,y)
plt.title("First Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
# Adding style to your graph
import matplotlib.pyplot as plt

x1 = [3,5,7]
y1 = [2,4,6]

x2 = [1, 4,7]
y2 = [2,5,1]

plt.plot(x1,y1, label="Line1",linewidth=5)
plt.plot(x2, y2, label="Line2", linewidth=5)

plt.title("Style Graph")
plt.xlabel("XAxis")
plt.ylabel("YAxis")
plt.legend()
plt.grid(True)
plt.show()
```

# Bar Graph
- Diffrence of groups over time period
- Used for catagarial data
- Example : speakers of different language

```python
import matplotlib.pyplot as plt

y1 = [1000, 2000, 3000, 5000]
x1 = ["Marathi", "Tamil", "Hindi", "English"]


plt.bar(x1, y1)

plt.title("Language Usage")
plt.xlabel("Languages")
plt.ylabel("Speakers")
plt.legend()
plt.show()
```

# Histogram
- A histogram represents the frequency distribution of continuous variables/quantitative variable

```python
import matplotlib.pyplot as plt

gdp = [50, 40, 30, 60, 80, 90]


plt.hist(gdp, bins="auto", histtype="bar", rwidth=1)
plt.show()
```
